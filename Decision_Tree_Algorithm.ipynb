# Data Preprocessing

```python
x=df.drop(columns=['Species','Id'])
y=df['Species']
```

```python
x.head()
```

```python
y.head()
```

# Splitting the data into Train and Test

```python
from sklearn.model_selection import train_test_split  
X_train, X_test, y_train, y_test = train_test_split(x, y,test_size=0.2, random_state=0,stratify=y) 
```

# Training the Model

Here, I trained the model using Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier
tree = DecisionTreeClassifier(criterion='gini', max_depth=4, random_state=0)
tree = tree.fit(X_train, y_train)
print('Decision Tree Classifer Created')
```

# Evaluating the Model

```python
y_pred = tree.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
print('Accuracy Score:',accuracy_score(y_test,y_pred))
```

```python
df=pd.DataFrame({"Actual_y":y_test ,"Predicted_y":y_pred})
df.head()
```

# Importing necessary libraries for graph visualization

```python
from pydotplus import graph_from_dot_data
from sklearn.tree import export_graphviz
from IPython.display import Image
```

# Visualizing the graph

```python
dot_data = export_graphviz(tree, 
                           filled=True, 
                           rounded=True,
                           class_names=['Iris-setosa', 'Iris-versicolor', 'Iris-virginica'],
                           feature_names=[ 'SepalLengthCm','SepalWidthCm' ,'PetalLengthCm' ,'PetalWidthCm'],
                           out_file=None)
graph = graph_from_dot_data(dot_data)
```

```python
Image(graph.create_png())
```

# Predicting New Data

```python
hours = [[4.3	,3	,1.1	,0.1]]
own_pred = tree.predict(hours)
print("RESULT:")
print(own_pred[0])
```
